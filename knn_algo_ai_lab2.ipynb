```python
# knn algorithm n featured
import math

class knn_class_algorithm:
    def creating_dictionary(self, age, income, education, experience, type_):
        data = {}
        for idx in range(len(age)):
            data[idx] = {
                'age': age[idx],
                'income': income[idx],
                'education': education[idx],
                'experience': experience[idx],
                'type_': type_[idx]
            }
        return data

    def print_data_in_table(self, data):
        counterA = 0
        counterB = 0
        counterC = 0
        print("SN\tAge\tIncome\tEducation\tExperience\tType")
        for key in data:
            print(f"{key}\t{data[key]['age']}\t{data[key]['income']}\t{data[key]['education']}\t\t{data[key]['experience']}\t\t{data[key]['type_']}")
            if data[key]['type_'] == 'A':
                counterA += 1
            elif data[key]['type_'] == 'B':
                counterB += 1
            else:
                counterC += 1
        print(f"Count -> A:{counterA}, B:{counterB}, C:{counterC}")

    def print_data_with_distances_using_provided_data(self, data, given_features):
        distances = {}
        for key in data:
            dist_sum = 0
            for feature in ['age', 'income', 'education', 'experience']:
                diff = given_features[feature] - data[key][feature]
                dist_sum += diff ** 2
            distance = math.sqrt(dist_sum)
            print(distance)
            distances[key] = {'distance': distance}
        print("SN\tAge\tIncome\tEducation\tExperience\tType\tDistance")
        for key in data:
            print(f"{key}\t{data[key]['age']}\t{data[key]['income']}\t{data[key]['education']}\t\t{data[key]['experience']}\t\t{data[key]['type_']}\t{distances[key]['distance']:.2f}")
        return distances

    def return_sorted_distances_data(self, distances):
        sorted_distances = dict(sorted(distances.items(), key=lambda x: x[1]['distance']))
        return sorted_distances

    def count_type_according_to_knn(self, data, sorted_distances, countmax):
        counterA = 0
        counterB = 0
        counterC = 0
        count = 0
        for key in sorted_distances:
            if data[key]['type_'] == 'A':
                counterA += 1
            elif data[key]['type_'] == 'B':
                counterB += 1
            elif data[key]['type_'] == 'C':
                counterC += 1
            count += 1
            if count == countmax:
                break
        return counterA, counterB, counterC

    def knn_result(self, counterA, counterB, counterC):
        print(f"Counts -> A:{counterA}, B:{counterB}, C:{counterC}")
        output = max(counterA, counterB, counterC)
        if output == counterA:
            type_selected = 'A'
        elif output == counterB:
            type_selected = 'B'
        else:
            type_selected = 'C'
        return type_selected

    def knn_algorithm(self, age, income, education, experience, type_, given_features, countmax):
        dictionary = self.creating_dictionary(age, income, education, experience, type_)
        self.print_data_in_table(dictionary)
        distances_dictionary = self.print_data_with_distances_using_provided_data(dictionary, given_features)
        sorted_distances_dictionary = self.return_sorted_distances_data(distances_dictionary)
        A, B, C = self.count_type_according_to_knn(dictionary, sorted_distances_dictionary, countmax)
        result = self.knn_result(A, B, C)
        return result


# Your original data
age = [22, 28, 35, 41, 30, 24, 46, 39, 21, 33,
       48, 29, 37, 50, 26, 32, 43, 25, 36, 44,
       27, 31, 47, 34, 23, 38, 42, 20, 45, 40]

income = [35500, 52100, 74200, 85600, 45000, 31800, 95200, 60000, 28300, 58400,
          112600, 49500, 71300, 120000, 41700, 59900, 89100, 38000, 68500, 93400,
          42600, 55800, 108900, 63200, 30600, 70000, 87700, 25400, 96300, 76100]

type_ = ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B', 'A', 'B',
         'C', 'A', 'B', 'A', 'A', 'B', 'C', 'A', 'B', 'B',
         'A', 'B', 'B', 'B', 'A', 'B', 'C', 'A', 'C', 'B']

# New features you want to add (education in years, experience in years)
education = [12, 16, 14, 18, 13, 12, 20, 16, 10, 15,
             20, 14, 16, 22, 11, 15, 19, 13, 17, 21,
             12, 16, 20, 15, 10, 18, 19, 9, 21, 17]

experience = [1, 5, 10, 15, 3, 2, 20, 12, 0, 7,
              21, 6, 8, 23, 3, 9, 15, 4, 10, 17,
              2, 7, 20, 10, 1, 12, 15, 0, 18, 11]

# Query point (new person)
given_features = {
    'age': 30,
    'income': 60000,
    'education': 16,
    'experience': 5
}

countmax = 5

knn = knn_class_algorithm()
result = knn.knn_algorithm(age, income, education, experience, type_, given_features, countmax)

print(f"\nAccording to KNN, person with features {given_features} is classified as type: {result}")
```

```python
# knn alogrithm 4 categorical data
import numpy as np
from collections import Counter

# Sample dataset with 4 numeric + 4 categorical features

age = [22, 28, 35, 41, 30, 24, 46, 39, 21, 33,
       48, 29, 37, 50, 26, 32, 43, 25, 36, 44,
       27, 31, 47, 34, 23, 38, 42, 20, 45, 40]

income = [35500, 52100, 74200, 85600, 45000, 31800, 95200, 60000, 28300, 58400,
          112600, 49500, 71300, 120000, 41700, 59900, 89100, 38000, 68500, 93400,
          42600, 55800, 108900, 63200, 30600, 70000, 87700, 25400, 96300, 76100]

education = [12, 16, 14, 18, 13, 12, 20, 16, 10, 15,
             20, 14, 16, 22, 11, 15, 19, 13, 17, 21,
             12, 16, 20, 15, 10, 18, 19, 9, 21, 17]

experience = [1, 5, 10, 15, 3, 2, 20, 12, 0, 7,
              21, 6, 8, 23, 3, 9, 15, 4, 10, 17,
              2, 7, 20, 10, 1, 12, 15, 0, 18, 11]

# Categorical features (4 of them)
gender = ['M', 'F', 'M', 'M', 'F', 'F', 'M', 'F', 'M', 'F',
          'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M',
          'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F']

city = ['NY', 'LA', 'NY', 'CHI', 'LA', 'CHI', 'NY', 'LA', 'CHI', 'NY',
        'LA', 'CHI', 'NY', 'LA', 'CHI', 'NY', 'LA', 'CHI', 'NY', 'LA',
        'CHI', 'NY', 'LA', 'CHI', 'NY', 'LA', 'CHI', 'NY', 'LA', 'CHI']

profession = ['Engineer', 'Doctor', 'Artist', 'Engineer', 'Artist', 'Doctor', 'Engineer', 'Doctor', 'Artist', 'Engineer',
              'Artist', 'Doctor', 'Engineer', 'Artist', 'Doctor', 'Engineer', 'Artist', 'Doctor', 'Engineer', 'Artist',
              'Doctor', 'Engineer', 'Artist', 'Doctor', 'Engineer', 'Artist', 'Doctor', 'Engineer', 'Artist', 'Doctor']

marital_status = ['Single', 'Married', 'Single', 'Married', 'Single', 'Married', 'Single', 'Married', 'Single', 'Married',
                  'Single', 'Married', 'Single', 'Married', 'Single', 'Married', 'Single', 'Married', 'Single', 'Married',
                  'Single', 'Married', 'Single', 'Married', 'Single', 'Married', 'Single', 'Married', 'Single', 'Married']

type_ = ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B', 'A', 'B',
         'C', 'A', 'B', 'A', 'A', 'B', 'C', 'A', 'B', 'B',
         'A', 'B', 'B', 'B', 'A', 'B', 'C', 'A', 'C', 'B']

# Normalize numeric features
def normalize(features):
    features = np.array(features)
    min_val = features.min()
    max_val = features.max()
    return (features - min_val) / (max_val - min_val)

age_norm = normalize(age)
income_norm = normalize(income)
education_norm = normalize(education)
experience_norm = normalize(experience)

# Combine numeric features as a list of lists
numeric_features = list(zip(age_norm, income_norm, education_norm, experience_norm))

# Categorical features combined
categorical_features = list(zip(gender, city, profession, marital_status))

def mixed_distance(point1_num, point2_num, point1_cat, point2_cat, weight_num=0.7, weight_cat=0.3):
    # Euclidean distance for numeric features
    num_dist = np.linalg.norm(np.array(point1_num) - np.array(point2_num))
    # Hamming distance for categorical features (sum of mismatches)
    cat_dist = sum(p1 != p2 for p1, p2 in zip(point1_cat, point2_cat)) / len(point1_cat)
    # Weighted sum
    return weight_num * num_dist + weight_cat * cat_dist

def knn_mixed(features_num, features_cat, labels, query_num, query_cat, k=5):
    distances = []
    for i in range(len(labels)):
        dist = mixed_distance(features_num[i], features_num[i], features_cat[i], features_cat[i], weight_num=0.7, weight_cat=0.3)
        dist = mixed_distance(features_num[i], query_num, features_cat[i], query_cat)
        distances.append((dist, labels[i]))
    distances.sort(key=lambda x: x[0])
    neighbors = distances[:k]
    votes = Counter(label for _, label in neighbors)
    return votes.most_common(1)[0][0]

# Example query
query_numeric = [30, 60000, 16, 5]
query_numeric_norm = [
    (30 - min(age)) / (max(age) - min(age)),
    (60000 - min(income)) / (max(income) - min(income)),
    (16 - min(education)) / (max(education) - min(education)),
    (5 - min(experience)) / (max(experience) - min(experience))
]

query_categorical = ('M', 'NY', 'Engineer', 'Single')

result = knn_mixed(numeric_features, categorical_features, type_, query_numeric_norm, query_categorical, k=5)
print(f"Predicted type for query: {result}")
```
